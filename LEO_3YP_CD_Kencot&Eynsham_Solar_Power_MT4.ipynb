```python
#analysis of energy output of Kencot Hill solar farm; and output per metre squared...

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime

#import data from Kencot Hill solar farm (estimated capacity 37MW)
df_kh = pd.read_csv("ninja_pv_kencot_hill_37000_kW_pd.csv", index_col=0,parse_dates=True,dayfirst=True, delimiter=",")
df_kh.head()
df_kh.plot()

df_kh_daily = df_kh.resample("D").sum()

df_kh_daily.plot()
```

```python
#import data from estimtaed Eynsham Park&Ride (estimated capacity of 2.16MW)
df_epr = pd.read_csv("ninja_pv_eynsham_p&r_2160_kW_pd.csv", index_col=0,parse_dates=True,dayfirst=True,delimiter=",")
df_epr.plot()
df_epr.head(20)
```

```python
#import data from estimated power output per m^2 in Eynsham area (estimated capacity of 90W)

df_squaremetre = pd.read_csv("ninja_pv_eynsham_p&r_0.09_kW_pd.csv", index_col=0,parse_dates=True,dayfirst=True,delimiter=",")
df_squaremetre.plot()
```

```python
#display daily output energy for kencot
df_kh_day = df_kh.resample("D").sum()
df_kh_day.plot()
```

```python
#estimate annual output from Kencot Hill Solar Farm

df_kh_year = df_kh.resample("Y").sum()
kh_year_output = df_kh_year.at["2019-12-31","electricity (kWh)"]
print("Yearly output from Kencot Hill Solar farm:",kh_year_output,"kWh")
```

```python
#estimate annual output from 2.16MW capacity installed at Eynsham Park & Ride

df_epr_year = df_epr.resample("Y").sum()
epr_year_output = df_epr_year.at["2019-12-31","electricity (kWh)"]
print("Yearly output from Eynsham P&R Solar farm:",epr_year_output,"kWh")

#estimate max daily output from 2.16MW capacity installed at Eynsham Park & Ride

df_epr_day = df_epr.resample("D").sum()

df_epr_day_max = df_epr_day.max()
print("Max daily output from Eynsham P&R",df_epr_day_max,"kWh")
df_epr_day.head()
```

```python
#estimate how much annual output provided per metre-squared of solar installed at stations, rooftops etc

df_squaremetre_year = df_squaremetre.resample("Y").sum()
squaremetre_year_output = df_squaremetre_year.at["2019-12-31","electricity (kWh)"]
print("Yearly output from 1 metre squared of solar farm",squaremetre_year_output,"kWh")
```
